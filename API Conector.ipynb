```
import requests
import json
id=123456
for x in range(4):
  r = requests.get('https://priscila-moraes1.outsystemscloud.com/Drug_Thru/rest/API_Robot_Drug_Thru/Get_pedido?Id_nota_venda='+str(id+x))
  parsed = json.loads(r.text)
  print('Nota de venda: '+(str(parsed[0]["NotaVendaId"])))
  print('Qtd. de Itens: '+(str(len(parsed)))+'\n')
  for y in range(len(parsed)):
    #print('N° da Ordem: '+(str(parsed[y]["Id_order_nota_venda"])))
    print('Cód. Produto: '+(str(parsed[y]["Cod_produto"])))
    print('Produto '+str(y+1)+': '+parsed[y]["Item_nota_venda"]+'\n')
    
  print('----------------------------------------------------------------------------------------\n')
  
```
